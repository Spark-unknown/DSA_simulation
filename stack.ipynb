```python
import pygame
import sys

# Constants
WINDOW_WIDTH = 800
WINDOW_HEIGHT = 600
STACK_WIDTH = 200
STACK_HEIGHT = 400
ITEM_WIDTH = 150
ITEM_HEIGHT = 40
BACKGROUND_COLOR = (255, 255, 255)
STACK_COLOR = (200, 200, 200)
ITEM_COLOR = (100, 200, 100)
TEXT_COLOR = (0, 0, 0)

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Stack Visualization")
font = pygame.font.Font(None, 36)

# Stack Initialization
stack = []
max_size = STACK_HEIGHT // ITEM_HEIGHT

def draw_stack(stack):
    # Draw the stack rectangle
    pygame.draw.rect(screen, STACK_COLOR, (300, 100, STACK_WIDTH, STACK_HEIGHT), 0)
    
    # Draw each item in the stack
    for index, item in enumerate(reversed(stack)):  # Reverse to display the top of the stack at the top
        item_x = 305  # x position of the item
        item_y = 100 + index * ITEM_HEIGHT  # y position of the item
        pygame.draw.rect(screen, ITEM_COLOR, (item_x, item_y, ITEM_WIDTH, ITEM_HEIGHT), 0)
        text_surface = font.render(str(item), True, TEXT_COLOR)
        screen.blit(text_surface, (item_x + 10, item_y + 10))  # Center text in item

def push():
    if len(stack) < max_size:
        new_item = len(stack) + 1  # Create a new item based on the current stack size
        stack.append(new_item)

def pop():
    if stack:
        stack.pop()

# Game Loop
while True:
    screen.fill(BACKGROUND_COLOR)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_p:  # Push on 'P' key press
                push()
            elif event.key == pygame.K_o:  # Pop on 'O' key press
                pop()

    draw_stack(stack)
    
    # Display instructions
    instruction_surface = font.render("Press 'P' to Push, 'O' to Pop", True, TEXT_COLOR)
    screen.blit(instruction_surface, (300, 20))

    pygame.display.flip()
    pygame.time.delay(100)

```

```python
import pygame
import sys
from collections import deque

# Constants
WINDOW_WIDTH = 800
WINDOW_HEIGHT = 600
QUEUE_WIDTH = 600
QUEUE_HEIGHT = 400
ITEM_WIDTH = 50
ITEM_HEIGHT = 40
BACKGROUND_COLOR = (255, 255, 255)
QUEUE_COLOR = (200, 200, 200)
ITEM_COLOR = (100, 200, 100)
TEXT_COLOR = (0, 0, 0)

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Queue Visualization")
font = pygame.font.Font(None, 36)

# Queue Initialization
queue = deque()
max_size = QUEUE_WIDTH // ITEM_WIDTH

def draw_queue(queue):
    # Draw the queue rectangle
    pygame.draw.rect(screen, QUEUE_COLOR, (100, 100, QUEUE_WIDTH, QUEUE_HEIGHT), 0)
    
    # Draw each item in the queue
    for index, item in enumerate(queue):
        item_x = 100 + index * ITEM_WIDTH + 10  # x position of the item
        item_y = 120  # y position of the item
        pygame.draw.rect(screen, ITEM_COLOR, (item_x, item_y, ITEM_WIDTH, ITEM_HEIGHT), 0)
        text_surface = font.render(str(item), True, TEXT_COLOR)
        screen.blit(text_surface, (item_x + 15, item_y + 10))  # Center text in item

def enqueue():
    if len(queue) < max_size:
        new_item = len(queue) + 1  # Create a new item based on the current queue size
        queue.append(new_item)

def dequeue():
    if queue:
        queue.popleft()

# Game Loop
while True:
    screen.fill(BACKGROUND_COLOR)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_e:  # Enqueue on 'E' key press
                enqueue()
            elif event.key == pygame.K_d:  # Dequeue on 'D' key press
                dequeue()

    draw_queue(queue)
    
    # Display instructions
    instruction_surface = font.render("Press 'E' to Enqueue, 'D' to Dequeue", True, TEXT_COLOR)
    screen.blit(instruction_surface, (100, 20))

    pygame.display.flip()
    pygame.time.delay(100)

```

```python
import pygame
import sys

# Constants
WINDOW_WIDTH = 800
WINDOW_HEIGHT = 600
LINKED_LIST_WIDTH = 600
LINKED_LIST_HEIGHT = 400
NODE_WIDTH = 100
NODE_HEIGHT = 40
BACKGROUND_COLOR = (255, 255, 255)
NODE_COLOR = (100, 200, 100)
TEXT_COLOR = (0, 0, 0)

# Node Class
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

# Linked List Class
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def pop(self):
        if not self.head:
            return None
        if not self.head.next:
            self.head = None
            return
        current = self.head
        while current.next.next:
            current = current.next
        current.next = None

    def remove_first(self):
        if self.head:
            self.head = self.head.next

    def to_list(self):
        result = []
        current = self.head
        while current:
            result.append(current.value)
            current = current.next
        return result

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Linked List Visualization")
font = pygame.font.Font(None, 36)

# Linked List Initialization
linked_list = LinkedList()

def draw_linked_list(linked_list):
    # Draw linked list nodes
    current = linked_list.head
    x, y = 100, 120
    while current:
        pygame.draw.rect(screen, NODE_COLOR, (x, y, NODE_WIDTH, NODE_HEIGHT), 0)
        text_surface = font.render(str(current.value), True, TEXT_COLOR)
        screen.blit(text_surface, (x + 10, y + 10))  # Center text in node
        x += NODE_WIDTH + 10  # Move x position for next node
        current = current.next

def draw_arrow(x, y):
    pygame.draw.polygon(screen, TEXT_COLOR, [(x + NODE_WIDTH, y + NODE_HEIGHT // 2), (x + NODE_WIDTH + 10, y + NODE_HEIGHT // 2 - 5), (x + NODE_WIDTH + 10, y + NODE_HEIGHT // 2 + 5)])

# Game Loop
while True:
    screen.fill(BACKGROUND_COLOR)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_a:  # Add item on 'A' key press
                linked_list.append(len(linked_list.to_list()) + 1)  # Add next number
            elif event.key == pygame.K_r:  # Remove last item on 'R' key press
                linked_list.pop()
            elif event.key == pygame.K_d:  # Remove first item on 'D' key press
                linked_list.remove_first()

    draw_linked_list(linked_list)
    
    # Draw arrows to represent links
    current = linked_list.head
    x, y = 100, 120
    while current and current.next:
        draw_arrow(x, y)
        x += NODE_WIDTH + 10
        current = current.next
    
    # Display instructions
    instruction_surface = font.render("Press 'A' to Add Item, 'R' to Remove Last Item, 'D' to Remove First Item", True, TEXT_COLOR)
    screen.blit(instruction_surface, (100, 20))

    pygame.display.flip()
    pygame.time.delay(100)
```

```python
import pygame
import sys

# Constants
BACKGROUND_COLOR = (255, 255, 255)
NODE_COLOR = (100, 200, 100)
TEXT_COLOR = (0, 0, 0)

# Node Class for Doubly Linked List
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None

# Doubly Linked List Class
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, value):
        new_node = Node(value)
        if not self.head:  # If the list is empty
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node

    def prepend(self, value):
        new_node = Node(value)
        if not self.head:  # If the list is empty
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def pop(self):
        if self.tail:
            if self.tail.prev:
                self.tail = self.tail.prev
                self.tail.next = None
            else:  # List has only one node
                self.head = None
                self.tail = None

    def remove_first(self):
        if self.head:
            if self.head.next:
                self.head = self.head.next
                self.head.prev = None
            else:  # List has only one node
                self.head = None
                self.tail = None

    def to_list(self):
        result = []
        current = self.head
        while current:
            result.append(current.value)
            current = current.next
        return result

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
pygame.display.set_caption("Responsive Doubly Linked List Visualization")
font = pygame.font.Font(None, 36)

# Doubly Linked List Initialization
doubly_linked_list = DoublyLinkedList()

def draw_doubly_linked_list(linked_list, window_width):
    current = linked_list.head
    node_width = window_width // 15  # Dynamic node width
    node_height = 40

    # Center the list in the window
    total_nodes = len(linked_list.to_list())
    total_width = (node_width + 10) * total_nodes - 10  # total width of the nodes with spacing
    start_x = (window_width - total_width) // 2  # Centering start position

    y = 120
    while current:
        pygame.draw.rect(screen, NODE_COLOR, (start_x, y, node_width, node_height), 0)
        text_surface = font.render(str(current.value), True, TEXT_COLOR)
        screen.blit(text_surface, (start_x + 10, y + 10))  # Center text in node
        start_x += node_width + 10  # Move x position for next node
        current = current.next

def draw_arrow(x, y, direction):
    if direction == 'right':
        pygame.draw.polygon(screen, TEXT_COLOR, [(x + 100, y + 20), (x + 110, y + 15), (x + 110, y + 25)])
    elif direction == 'left':
        pygame.draw.polygon(screen, TEXT_COLOR, [(x, y + 20), (x - 10, y + 15), (x - 10, y + 25)])

# Game Loop
while True:
    screen.fill(BACKGROUND_COLOR)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_a:  # Add item on 'A' key press
                doubly_linked_list.append(len(doubly_linked_list.to_list()) + 1)  # Add next number
            elif event.key == pygame.K_p:  # Prepend item on 'P' key press
                doubly_linked_list.prepend(len(doubly_linked_list.to_list()) + 1)  # Add next number
            elif event.key == pygame.K_r:  # Remove last item on 'R' key press
                doubly_linked_list.pop()
            elif event.key == pygame.K_d:  # Remove first item on 'D' key press
                doubly_linked_list.remove_first()

    # Get the current window size
    window_width, window_height = screen.get_size()

    draw_doubly_linked_list(doubly_linked_list, window_width)
    
    # Draw arrows to represent links
    current = doubly_linked_list.head
    x = (window_width - (window_width // 15 + 10) * len(doubly_linked_list.to_list())) // 2  # Starting x
    y = 120
    while current and current.next:
        draw_arrow(x, y, 'right')
        x += (window_width // 15 + 10)
        current = current.next
    
    current = doubly_linked_list.tail
    x = (window_width - (window_width // 15 + 10) * len(doubly_linked_list.to_list())) // 2 + (window_width // 15 + 10) * (len(doubly_linked_list.to_list()) - 1)  # Starting x
    while current and current.prev:
        draw_arrow(x, y, 'left')
        x -= (window_width // 15 + 10)
        current = current.prev
    
    # Display instructions
    instruction_surface = font.render("Press 'A' to Append, 'P' to Prepend, 'R' to Remove Last, 'D' to Remove First", True, TEXT_COLOR)
    screen.blit(instruction_surface, (100, 20))

    pygame.display.flip()
    pygame.time.delay(100)
```

```python
import pygame
import sys

# Constants
WINDOW_WIDTH = 1000
WINDOW_HEIGHT = 600
NODE_RADIUS = 20
NODE_COLOR = (100, 200, 100)
EDGE_COLOR = (0, 0, 0)
TEXT_COLOR = (0, 0, 0)
BACKGROUND_COLOR = (255, 255, 255)

# Graph Class
class Graph:
    def __init__(self):
        self.nodes = {}  # {node_value: position}
        self.edges = []  # [(start_value, end_value)]

    def add_node(self, value, position):
        self.nodes[value] = position

    def remove_node(self, value):
        if value in self.nodes:
            del self.nodes[value]
            # Remove all edges associated with the node
            self.edges = [edge for edge in self.edges if value not in edge]

    def add_edge(self, start, end):
        if start in self.nodes and end in self.nodes:
            self.edges.append((start, end))

    def remove_edge(self, start, end):
        if (start, end) in self.edges:
            self.edges.remove((start, end))

    def to_list(self):
        return list(self.nodes.keys())

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Graph Visualization")
font = pygame.font.Font(None, 36)

# Graph Initialization
graph = Graph()

def draw_graph(graph):
    # Draw edges
    for start, end in graph.edges:
        start_pos = graph.nodes[start]
        end_pos = graph.nodes[end]
        pygame.draw.line(screen, EDGE_COLOR, start_pos, end_pos, 2)

    # Draw nodes
    for value, position in graph.nodes.items():
        pygame.draw.circle(screen, NODE_COLOR, position, NODE_RADIUS)
        text_surface = font.render(str(value), True, TEXT_COLOR)
        screen.blit(text_surface, (position[0] - NODE_RADIUS // 2, position[1] - NODE_RADIUS // 2))  # Center text in node

# Main Loop
while True:
    screen.fill(BACKGROUND_COLOR)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_n:  # Add node on 'N' key press
                node_value = len(graph.to_list()) + 1
                position = (50 + (node_value - 1) * 100, 50)  # Horizontal layout
                graph.add_node(node_value, position)

            elif event.key == pygame.K_e:  # Add edge on 'E' key press
                nodes_list = graph.to_list()
                if len(nodes_list) >= 2:
                    graph.add_edge(nodes_list[-2], nodes_list[-1])  # Connect last two added nodes

            elif event.key == pygame.K_r:  # Remove last added node
                nodes_list = graph.to_list()
                if nodes_list:
                    graph.remove_node(nodes_list[-1])

            elif event.key == pygame.K_d:  # Remove last added edge
                if len(graph.edges) > 0:
                    graph.edges.pop()  # Remove the last edge

    draw_graph(graph)
    
    # Display instructions
    instruction_surface = font.render(
        "Press 'N' to Add Node, 'E' to Add Edge, 'R' to Remove Node, 'D' to Remove Edge",
        True,
        TEXT_COLOR
    )
    screen.blit(instruction_surface, (20, 20))

    pygame.display.flip()
    pygame.time.delay(100)
```

```python
import pygame
import sys

# Constants
WINDOW_WIDTH = 800
WINDOW_HEIGHT = 600
NODE_RADIUS = 20
NODE_COLOR = (100, 200, 100)
EDGE_COLOR = (0, 0, 0)
TEXT_COLOR = (0, 0, 0)
BACKGROUND_COLOR = (255, 255, 255)
HORIZONTAL_SPACING = 100
VERTICAL_SPACING = 60

# Node class
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Binary Tree class
class BinaryTree:
    def __init__(self):
        self.root = None
        self.nodes = []
        self.selected_node = None  # To keep track of the currently selected node

    def insert(self, value, parent=None):
        if self.root is None:
            self.root = Node(value)
            self.nodes.append(self.root)
            self.selected_node = self.root  # Select the root node
        else:
            if parent is None:
                parent = self.selected_node
            if parent is not None:
                if value < parent.value:
                    if parent.left is None:
                        parent.left = Node(value)
                        self.nodes.append(parent.left)
                        self.selected_node = parent.left  # Select the newly added node
                else:
                    if parent.right is None:
                        parent.right = Node(value)
                        self.nodes.append(parent.right)
                        self.selected_node = parent.right  # Select the newly added node

    def draw(self, surface, node, x, y, level=0):
        if node is not None:
            # Draw node
            pygame.draw.circle(surface, NODE_COLOR, (x, y), NODE_RADIUS)
            text_surface = font.render(str(node.value), True, TEXT_COLOR)
            surface.blit(text_surface, (x - NODE_RADIUS // 2, y - NODE_RADIUS // 2))

            # Calculate positions for children
            child_x = x - (HORIZONTAL_SPACING // 2) * (2 ** (1 + level))
            child_y = y + VERTICAL_SPACING

            # Draw left child
            if node.left is not None:
                pygame.draw.line(surface, EDGE_COLOR, (x, y), (child_x + HORIZONTAL_SPACING // 2, child_y), 2)
                self.draw(surface, node.left, child_x + HORIZONTAL_SPACING // 2, child_y, level + 1)

            # Draw right child
            child_x = x + (HORIZONTAL_SPACING // 2) * (2 ** (1 + level))
            if node.right is not None:
                pygame.draw.line(surface, EDGE_COLOR, (x, y), (child_x - HORIZONTAL_SPACING // 2, child_y), 2)
                self.draw(surface, node.right, child_x - HORIZONTAL_SPACING // 2, child_y, level + 1)

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Binary Tree Visualization")
font = pygame.font.Font(None, 36)

# Binary Tree Initialization
binary_tree = BinaryTree()

# Main Loop
while True:
    screen.fill(BACKGROUND_COLOR)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_a:  # Add node with user-defined value
                node_value = input("Enter the value for the new node: ")
                try:
                    node_value = int(node_value)
                    binary_tree.insert(node_value)
                    print(f"Added node with value: {node_value}")
                except ValueError:
                    print("Invalid input! Please enter a valid integer.")
            if event.key == pygame.K_l:  # Add left child
                if binary_tree.selected_node is not None:
                    node_value = input("Enter the value for the new left child: ")
                    try:
                        node_value = int(node_value)
                        binary_tree.insert(node_value, binary_tree.selected_node)
                        print(f"Added left child with value: {node_value}")
                    except ValueError:
                        print("Invalid input! Please enter a valid integer.")
            if event.key == pygame.K_r:  # Add right child
                if binary_tree.selected_node is not None:
                    node_value = input("Enter the value for the new right child: ")
                    try:
                        node_value = int(node_value)
                        binary_tree.insert(node_value, binary_tree.selected_node)
                        print(f"Added right child with value: {node_value}")
                    except ValueError:
                        print("Invalid input! Please enter a valid integer.")
            if event.key == pygame.K_n:  # Select last added node as parent
                if binary_tree.nodes:
                    binary_tree.selected_node = binary_tree.nodes[-1]  # Select the last added node
                    print(f"Selected node with value: {binary_tree.selected_node.value}")

    # Draw the binary tree
    if binary_tree.root is not None:
        binary_tree.draw(screen, binary_tree.root, WINDOW_WIDTH // 2, 50)

    # Display instructions
    instruction_surface = font.render("Press A to Add Node, L for Left Child, R for Right Child, N to Select Last Node", True, TEXT_COLOR)
    screen.blit(instruction_surface, (20, 20))

    pygame.display.flip()
    pygame.time.delay(100)
```

```python
import pygame
import sys
import random

# Constants
WINDOW_WIDTH, WINDOW_HEIGHT = 800, 600
BACKGROUND_COLOR = (255, 255, 255)
TEXT_COLOR = (0, 0, 0)
FONT_SIZE = 24
NODE_COLOR = (100, 200, 100)
FOUND_COLOR = (255, 0, 0)
CURRENT_COLOR = (255, 255, 0)

# Initialize Pygame and mixer
pygame.init()
pygame.mixer.init()

# Set up display
screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Array Search Visualization")
font = pygame.font.Font(None, FONT_SIZE)

# Load sound files
found_sound = pygame.mixer.Sound("found_sound.wav")  # Replace with your sound file path
not_found_sound = pygame.mixer.Sound("not_found_sound.wav")  # Replace with your sound file path

# Initialize variables
array = []
search_value = ""
search_active = False
search_mode = None
search_index = -1
step_index = 0
search_steps = []

def add_random_element():
    """Add a random element to the array and sort it."""
    element = random.randint(1, 100)
    array.append(element)
    array.sort()  # Sort the array after adding a new element

def linear_search(target):
    """Perform linear search and return steps."""
    steps = []
    for index, value in enumerate(array):
        steps.append((index, value))
        if value == target:
            return index, steps
    return -1, steps

def binary_search(target):
    """Perform binary search on a sorted array and return steps."""
    left, right = 0, len(array) - 1
    steps = []
    while left <= right:
        mid = (left + right) // 2
        steps.append((mid, array[mid]))
        if array[mid] == target:
            return mid, steps
        elif array[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1, steps

def display_instructions():
    """Display instructions on the screen."""
    instruction_text = font.render("Press A to Add, L for Linear Search, B for Binary Search", True, TEXT_COLOR)
    screen.blit(instruction_text, (20, 20))

def display_array():
    """Render the array elements on the screen."""
    for i, value in enumerate(array):
        color = NODE_COLOR
        if i == search_index:
            color = FOUND_COLOR
        if search_steps and step_index < len(search_steps):
            current_index, _ = search_steps[step_index]
            if i == current_index:
                color = CURRENT_COLOR

        pygame.draw.rect(screen, color, (50 + i * 50, 100, 40, 40))
        value_text = font.render(str(value), True, BACKGROUND_COLOR)
        screen.blit(value_text, (50 + i * 50 + 10, 110))

def main_loop():
    """Main event loop for the Pygame application."""
    global search_active, search_value, search_mode, search_index, step_index, search_steps

    while True:
        screen.fill(BACKGROUND_COLOR)
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()

            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_a:
                    add_random_element()
                elif event.key == pygame.K_l and not search_active:
                    search_mode = "linear"
                    search_active = True
                elif event.key == pygame.K_b and not search_active:
                    search_mode = "binary"
                    search_active = True
                
                if search_active:
                    handle_search_input(event)

        display_instructions()
        display_array()

        if search_active:
            input_text = font.render(f"Enter value to search: {search_value}", True, TEXT_COLOR)
            screen.blit(input_text, (20, 160))
            if search_steps and step_index < len(search_steps):
                current_index, current_value = search_steps[step_index]
                step_text = font.render(f"Checking index {current_index}: {current_value}", True, TEXT_COLOR)
                screen.blit(step_text, (20, 200))
                step_index += 1  # Proceed to the next step

        pygame.display.flip()
        pygame.time.delay(100)

def handle_search_input(event):
    """Handle input for search value."""
    global search_value, search_index, search_active, step_index, search_steps

    if event.unicode.isdigit():
        search_value += event.unicode
    elif event.key == pygame.K_BACKSPACE:
        search_value = search_value[:-1]
    elif event.key == pygame.K_RETURN:
        if search_value:
            target = int(search_value)
            if search_mode == "linear":
                search_index, search_steps = linear_search(target)
            elif search_mode == "binary":
                search_index, search_steps = binary_search(target)

            if search_index != -1:
                print(f"{search_mode.capitalize()} Search: Element {target} found at index {search_index}.")
                found_sound.play()
            else:
                print(f"{search_mode.capitalize()} Search: Element {target} not found.")
                not_found_sound.play()
                
            search_value = ""
            search_active = False
            step_index = 0

if __name__ == "__main__":
    main_loop()
```
